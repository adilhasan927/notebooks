```python
def fix_central_bias(x):
    x = x.copy()
    m, n = x.shape[:2]
    cent_m, cent_n = m // 2, n // 2
    for i in range(m):
        for j in range(n):
            overlap_size = (m-abs(cent_m-i)) * (n-abs(cent_n-j))
            x[i,j] /= overlap_size
    return x
```

```python
def downscale(image, patchsize):
    # Image shape: (rows, cols, bands)
    # We reshape to: (rows, cols, bands, patchrows, patchcols)
    image = np.pad(image, ((0, (patchsize - image.shape[0] % patchsize) if image.shape[0] % patchsize != 0 else 0),
                           (0, (patchsize - image.shape[1] % patchsize) if image.shape[0] % patchsize != 0 else 0),
                           (0,0)),
                   constant_values=((float('nan'), float('nan')),)*3)
    image = rearrange(image,
                      '(rows patchrows) (cols patchcols) bands -> rows cols bands patchrows patchcols',
                      patchrows=patchsize, patchcols=patchsize)
    image = np.nanmean(image, axis=(-1, -2))
    return image
```

## Produce our embeddings tifs

```python
list(downscales.keys())
```

```python
#downscales_embeds = dict()
for key, img in list(downscales.items())[-4:-1]:
    #if key[0] != 0:
    #    continue
    
    img = img[:,:,channels]
    img = rearrange(img, '(b t h) w c -> b c t h w', b=1, t=1)
    
    original_h, original_w = img.shape[-2:]
    pad_h = 224 - (original_h % 224)
    pad_w = 224 - (original_w % 224)
    pad_h = pad_h % 224
    pad_w = pad_w % 224
    img = np.pad(img, ((0, 0), (0, 0), (0, 0), (0, pad_h), (0, pad_w)), mode='reflect')

    img = torch.Tensor(img)
    img_ = img
    img = img.unfold(3, 224, 224).unfold(4, 224, 224)

    post_fold_shape = img.shape
    img = rearrange(img, 'b c t h1 w1 h w -> (b h1 w1) c t h w', h=224, w=224)
    
    batch_size = 1
    num_batches = math.ceil(img.shape[0] / batch_size) if img.shape[0] > batch_size else 1
    img = torch.tensor_split(img, num_batches, dim=0)
    
    rec_imgs = []
    for x in tqdm(img):
        with torch.no_grad():
            gc.collect()
            
            # At this point: [16, 6, 1, 224, 224]
            # We have to rearrange to: [16, 6, 1, 14, 14, 16, 16]
            x = x.to(device)
            x = rearrange(x, 'b c t (h1 h) (w1 w) -> b c t h1 w1 h w',
                          h=16, w=16)

            # Now we have: [16, 6, 1, 14, 14, 16, 16]
            # We have to get the ones where the last two axii contain an invalid value

            x_invalid = torch.isnan(x)
            x_invalid = x_invalid.any(-1).any(-1).any(1).any(1)
            # The shape of x_invalid: [16, 14, 14]

            # Check: is x_invalid all-False?
            # Then, we just return zeros of the correct shape.
            if x_invalid.flatten().all():
                rec_img = torch.zeros((1, 196, 768))
                rec_imgs.append(rec_img.cpu())
                #print('skip')
                continue

            # Now, we need to bring x back to [16, 6, 1, 224, 224]
            x = rearrange(x, 'b c t h1 w1 h w -> b c t (h1 h) (w1 w)',
              h=16, w=16)

            # Now, we do the patch embedding
            x_1 = model.patch_embed(x) + model.pos_embed[:,1:,:]
            
            # At this point, our shape: [16, 196, 768]
            # We need to make x_invalid broadcast against it.
            x_invalid = rearrange(x_invalid, 'b h w -> b (h w)')

            # x_invalid will broadcast against this now
            x_1 = x_1[~x_invalid]

            # Now, we just copy the internals of model.forward_encoder
            # The shape of x: [_, 768]
            x_1 = rearrange(x_1, '(b hw) e -> b (hw) e', b=1)
            x = x_1

            # append cls token
            cls_token = model.cls_token + model.pos_embed[:, :1, :]
            cls_tokens = cls_token.expand(x.shape[0], -1, -1)
            x = torch.cat((cls_tokens, x), dim=1)

            # apply Transformer blocks
            for blk in model.blocks:
                x = blk(x)
            x = model.norm(x)

            # Shape of x now: [1, _+1, 768]
            # We need to: first, discard the CLS token
            # Then: replace the masked blocks with the zero vector
            x = x[:,1:]

            list_ = []
            j = 0
            for i in x_invalid.flatten():
                if not i:
                    list_.append(x[0,j])
                    j += 1
                else:
                    list_.append(torch.zeros(x[0,0].shape, device=device))
            x = torch.stack(list_).reshape((1, 196, 768))
            
            rec_img = x

            # Now, we have: [1, 196, 768]
            rec_imgs.append(rec_img.cpu())

    rec_imgs = torch.concat(rec_imgs, dim=0)

    #print('concatted')

    post_pad_h = original_h + pad_h
    post_pad_w = original_w + pad_w

    rec_imgs = rearrange(rec_imgs,
                         '(b h1 w1) (h w) e -> b h1 w1 h w e',
                         b=1,
                         h1=post_fold_shape[3], w1=post_fold_shape[4],
                         h=14, w=14, e=768)

    #print('rearrange1')
    
    rec_imgs = rearrange(rec_imgs,
                         'b h1 w1 h w e -> b (h1 h) (w1 w) e')

    #print('rearrange3')

    # At this point, we still have the padding...
    # pad_h, pad_w: 0 to 224
    # We want to turn this into: 0 to 16
    # And also, remove a conservative amount of padding
    pad_h_ = pad_h // 16
    pad_w_ = pad_w // 16

     Now, we remove the padding from the ends
    if pad_h_ > 0:
        rec_imgs = rec_imgs[:, :-pad_h_]
    if pad_w_ > 0:
        rec_imgs = rec_imgs[:, :, :-pad_w_]
    
    downscales_embeds[key] = rec_imgs.squeeze()
```
